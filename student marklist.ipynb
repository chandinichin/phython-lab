```python
class Student:
    def __init__(self, name, reg_number, course):
        self.name = name
        self.reg_number = reg_number
        self.course = course
        self.marks = []

    def input_marks(self):
        print(f"Enter marks for {self.name} ({self.reg_number}) in 5 subjects:")
        for i in range(5):
            mark = float(input(f"Subject {i + 1} mark: "))
            self.marks.append(mark)

    def calculate_total(self):
        return sum(self.marks)

    def calculate_average(self):
        return sum(self.marks) / len(self.marks)

    def calculate_grade(self):
        average = self.calculate_average()
        if average >= 90:
            return 'A'
        elif 80 <= average < 90:
            return 'B'
        elif 70 <= average < 80:
            return 'C'
        elif 60 <= average < 70:
            return 'D'
        else:
            return 'F'

def main():
    students = []

    while True:
        print("\nMenu:")
        print("1. Add Student")
        print("2. Input Marks")
        print("3. Display Details")
        print("4. Quit")

        choice = input("Enter your choice: ")

        if choice == '1':
            name = input("Enter student name: ")
            reg_number = input("Enter register number: ")
            course = input("Enter course: ")
            student = Student(name, reg_number, course)
            students.append(student)
            print(f"{name} added successfully!")

        elif choice == '2':
            reg_number = input("Enter register number of the student: ")
            student = next((s for s in students if s.reg_number == reg_number), None)
            if student:
                student.input_marks()
                print("Marks input successful!")
            else:
                print("Student not found!")

        elif choice == '3':
            reg_number = input("Enter register number of the student: ")
            student = next((s for s in students if s.reg_number == reg_number), None)
            if student:
                print("\nStudent Details:")
                print(f"Name: {student.name}")
                print(f"Register Number: {student.reg_number}")
                print(f"Course: {student.course}")
                print(f"Total Marks: {student.calculate_total()}")
                print(f"Average Marks: {student.calculate_average():.2f}")
                print(f"Grade: {student.calculate_grade()}")
            else:
                print("Student not found!")

        elif choice == '4':
            print("Exiting program. Goodbye!")
            break

        else:
            print("Invalid choice. Please enter a valid option.")

if __name__ == "__main__":
    main()

```
